# Ejercicio 1 - Resolver una EDO simple

```python
import numpy as np

def f(x,y): #Define la  ecuacion diferencial
  return x + y #Retorna el valor de dy/dx

x = 0 #Valor inicial de x
y = 1 #Condicion inicial y(0)=1

h = 0.1 #tamaño del paso

for i in range(5): #Ejecuta 5 iteraciones
  y = y + h*f(x,y) #aproximacion numerica tipo euler
  x = x + h #Avanza al siguiente punto

  print("x = ",x,"y = ",y) #Muestra resultados
```

# Ejercicio 2 - Visualización de la solución

```python
import matplotlib.pyplot as plt

def f(x,y): #Define la ecuacion diferencial
  return x + y #retorna pendiente

x = 0 #Valor inicial
y = 1 #Condicion inicial

h = 0.1 #Paso numerico

x_vals = [] #lista para almacenar valores de x
y_vals = [] #Lista para almacenar valores en y

for i in range(20): #Ejecuta 20 pasos
  x_vals.append(x) #guarda valor actual de x
  y_vals.append(y) #Guarda valor actual de y

  y = y + h * f(x,y) #Metodo de euler
  x = x + h #Incrementa en x

plt.plot(x_vals,y_vals) #Grafica la solucion aproximada
plt.xlabel("x") #Etiqueta eje x
plt.ylabel("y") #Etiqueta eje y
plt.title("Solucion aproximada de la EDO") #Titulo
plt.grid() #Activa cuadricula
plt.show() #Muestra grafica
```

# Ejercicio 3 - Aplicación IA : dinámica de aprendizaje

```python
import numpy as np

def f(t,w): #Define ecuacion diferencial
  return -0.5*w #Modelo de decaimento exponencial

t = 0 #Tiempo inicial

w = 10 #Paso inicial del modelo

h = 0.1 #Paso temporal

for i in range(20): #Ejecuta 20 iteraciones
  w = w + h*f(t,w) #Actualiza el peso
  t = t + h #Incrementa tiempo

  print("t = ",t,"w = ",w) #Muestra evolucion
```

# Ejercicio 2 - Comparación con solución exacta

```python
def f(x,y): #Define la ecuacion diferencial
  return x + y #retorna pendiente

def exacta(x): #Define la solucion analitica
  return 2*np.exp(x)-x-1 #formula exacta

x = 0 #Valor inicial
y = 1 #Condicion inicial

h = 0.1 #Paso numerico

for i in range(10): #Ejecuta 10 iteraciones
  y = y + h*f(x,y) #aproximacion numerica tipo euler
  x = x + h #Avanza al siguiente punto

  real = exacta(x) #Solucion exacta

  error = abs(real-y) #Calcula error absoluto

  print("x = ",x,"euler = ",y , "exacta = ",real, "Error: ",error) #Muestra resultados
```

# Ejercicio 1 - Método RK2

```python
import numpy as np

def f(x,y): #Define la  ecuacion diferencial
  return x + y #Retorna el valor de dy/dx

x = 0 #Valor inicial de x
y = 1 #Condicion inicial y(0)=1

h = 0.1 #tamaño del paso

n = 10

for i in range(n): #Ejecuta el metodo rk2
  k1 = f(x,y) #Calcula primer pendiente
  k2 = f(x + h/2 , y + h*k1/2) #Calcula segunda pendiente
  y = y + h*k2 #aproximacion numerica tipo euler
  x = x + h #Avanza al siguiente punto

  print("x = ",x,"y = ",y) #Muestra resultados
```

# Ejercicio 3 - Comparación Euler vs Rk4

```python

def f(x,y): #Define la ecuacion diferencial
  return x + y #retorna pendiente

def exacta(x): #Define la solucion analitica
  return 2*np.exp(x)-x-1 #formula exacta

x = 0 #Valor inicial

h = 0.1 #Paso numerico

y_euler = 1 #Condicion inicial para euler

y_rk4 = 1 #condicion inicial para rk4

for i in range(10):

  #Metodo de euler
  y_euler = y_euler + h*f(x,y_euler) #aproximacion numerica tipo euler

  #Metodo RK4
  k1 = f(x,y) #Calcula primer pendiente
  k2 = f(x + h/2 , y + h*k1/2) #Calcula segunda pendiente
  k3 = f(x + h/2 , y + h*k2/2) #Calcula tercer pendiente
  k4 = f(x + h , y + h*k3) #Calcula cuarta pendiente

  y_rk4 = y_rk4 + (h/6)*(k1 + 2*k2 + 2*k3 + k4) #Actualiza rk4

  x = x+h

  real = exacta(x) #Calcula la solucion exacta

  error_euler = abs(real-y_euler) #Error euler
  error_rk4 = abs(real-y_rk4) #Error rk4

  print("Exacta: ",real)
  print("Euler: ",y_euler)
  print("RK4: ",y_rk4)
  print("Error euler: ",error_euler)
  print("Error RK4: ",error_rk4)
```
